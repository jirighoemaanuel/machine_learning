Importing dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the concrete strength dataset

```python
concrete_strength_dataframe = pd.read_csv('RASHEED ML.csv')
```

```python
print(concrete_strength_dataframe.head())
```

```python
print(concrete_strength_dataframe.describe())
```

```python
correlation = concrete_strength_dataframe.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```

```python
sns.heatmap(correlation, fmt='.1f', annot=True, cmap="coolwarm",
            vmin=-1, vmax=1, center=0)
```

```python
df_zero_temp = concrete_strength_dataframe[concrete_strength_dataframe['Temperature'] == 0]
correlation_zero_temp = df_zero_temp.corr()
sns.heatmap(correlation_zero_temp, annot=True,
            cmap="coolwarm", vmin=-1, vmax=1, center=0)
```
